```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_diabetes
```

```python
df=load_diabetes()
```

```python
df.data
```

```python
df.feature_names
```

```python
df.target
```

```python
data= pd.DataFrame(df.data,columns= df.feature_names)
data.head()
```

```python
data['medv']=df.target
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
x=data.iloc[:,:-1]
y=data.iloc[:,-1:]
```

```python
from sklearn.preprocessing import StandardScaler
sc_x=StandardScaler()
x=sc_x.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
xtrain.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
dtr=DecisionTreeClassifier()
dtr.fit(xtrain,ytrain)
```

```python
ypred=dtr.predict(xtest)
```

```python
ypred[:10]
```

```python
ytest
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(18,9))
plot_tree(dtr,feature_names=df.feature_names,filled=True)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ypred,ytest)
```

```python
from sklearn.ensemble import RandomForestRegressor
rand_reg=RandomForestRegressor(n_estimators=1000)
rand_reg.fit(xtrain,ytrain)
```

```python
yprednew=rand_reg.predict(xtest)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(yprednew,ytest)
```

```python
mean_squared_error(ypred,ytest)
```
